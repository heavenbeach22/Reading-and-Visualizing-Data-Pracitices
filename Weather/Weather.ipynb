#### 1. Đọc dữ liệu từ các tập tin và tạo dataframe dfWeather với tên cột tương ứng

```python
# Lấy danh sách tên các tệp trong thư mục
TempHighF = pd.read_csv('Weather Data/1_TempHighF.txt',delimiter = ',',header = None)
TempLowF = pd.read_csv('Weather Data/2_TempLowF.txt', delimiter=',', header = None)
HumidityHighPercent = pd.read_csv('Weather Data/3_HumidityHighPercent.txt', delimiter=',', header = None)
HumidityLowPercent = pd.read_csv('Weather Data/4_HumidityLowPercent.txt', delimiter=',', header = None)
VisibilityHighMiles = pd.read_csv('Weather Data/5_VisibilityHighMiles.txt', delimiter=',', header = None)
VisibilityLowMiles = pd.read_csv('Weather Data/6_VisibilityLowMiles.txt', delimiter=',', header = None)
WindMPH = pd.read_csv('Weather Data/7_WindMPH.txt', delimiter=',', header = None)
PrecipitationSumInches = pd.read_csv('Weather Data/8_PrecipitationSumInches.txt', delimiter=',', header = None)
```

```python
# Đọc từng tệp và tạo DataFrame từ nội dung của từng tệp
TempHighF = TempHighF.T
TempHighF.head(5)
TempLowF = TempLowF.T
TempLowF.head(5)
HumidityHighPercent = HumidityHighPercent.T
HumidityHighPercent.head(5)
HumidityLowPercent = HumidityLowPercent.T
HumidityLowPercent.head()
VisibilityHighMiles = VisibilityHighMiles.T
VisibilityHighMiles.head()
VisibilityLowMiles = VisibilityLowMiles.T
VisibilityLowMiles.head()
WindMPH = WindMPH.T
WindMPH.head()
PrecipitationSumInches = PrecipitationSumInches.T
PrecipitationSumInches.head()
# Hiển thị 5 dòng đầu của DataFrame
```

```python
TempHighF.columns = ['TempHighF']
TempLowF.columns = ['TempLowF']
HumidityHighPercent.columns = ['HumidityHighPercent']
HumidityLowPercent.columns = ['HumidityLowPercent']
VisibilityHighMiles.columns = ['VisibilityHighMiles']
VisibilityLowMiles.columns = ['VisibilityLowMiles']
WindMPH.columns = ['WindMPH']
PrecipitationSumInches.columns = ['PrecipitationSumInches']
dfWeather = pd.concat([
    TempHighF,
    TempLowF,
    HumidityHighPercent,
    HumidityLowPercent,
    VisibilityHighMiles,
    VisibilityLowMiles,
    WindMPH,
    PrecipitationSumInches
], axis=1)
dfWeather.head()
```

```python
# Chuyển kiểu dữ liệu của tất cả các cột trong dfWeather thành numeric
dfWeather = dfWeather.apply(pd.to_numeric, errors='coerce')
dfWeather
```

### 2. Tạo dataframe dfWeather2 lần lượt với các cột sau đây:
- 'TempHighC', 'TempLowC', 'HumidityHighPercent', 'HumidityLowPercent', 'VisibilityHighKMs', 'VisibilityLowKMs', 'WindKMPH', 'PrecipitationSumMM'

```python
dfWeather2 = dfWeather.copy()
dfWeather2.head()
```

##### 2.1  Chuyển đổi cột nhiệt độ cao nhất từ độ F sang độ C và lưu vào cột TempHighC trong dfWeather2 (làm tròn 2 số lẻ), hiển thị 5 giá trị đầu của TempHighC
    - công thức chuyển từ độ F sang C: C = (F − 32) × 5/9
    - ví dụ chuyển từ 80F sang C: C = (80 - 32) x 5/9 = 26.67C

```python
dfWeather2['TempHighF'] = (dfWeather2['TempHighF'] - 32) * 5/9
dfWeather2.rename(columns={'TempHighF': 'TempHighC'}, inplace=True)
dfWeather2.head()
```

#### 2.2 Chuyển đổi cột nhiệt độ thấp nhất từ độ F sang độ C và lưu vào cột TempLowC trong dfWeather2

```python
dfWeather2['TempLowF'] = (dfWeather2['TempLowF'] - 32) * 5/9
dfWeather2.rename(columns={'TempLowF': 'TempLowC'}, inplace=True)
dfWeather2.head()
```

##### 2.3 Tính nhiệt độ trung bình TempMediumC trong dfWeather2

```python
dfWeather2["TempMediumC"] = (dfWeather2["TempHighC"] + dfWeather2["TempLowC"]) / 2
dfWeather2.head()
```

##### 2.4 Tạo mới các cột độ ẩm HumidityHighPercent và HumidityLowPercent trong dfWeather2 bằng các cột HumidityHighPercent và HumidityLowPercent trong dfWeather

```python
dfWeather2["HumidityHighPercent"] = dfWeather["HumidityHighPercent"]
dfWeather2["HumidityLowPercent"] = dfWeather["HumidityLowPercent"]
dfWeather2.head()
```

##### 2.5 Chuyển đổi cột tầm nhìn xa từ miles sang kilomet và lưu vào cột VisibilityHighKMs trong dfWeather2
    - công thức chuyển từ Mile sang Kilomet: km = mile * 1.609344
    - ví dụ chuyển từ 10 mile sang km: km = 10 * 1.609344 = 16.09344 km

```python
dfWeather2['VisibilityHighMiles']= dfWeather2['VisibilityHighMiles'] * 1.609344
dfWeather2.rename(columns={'VisibilityHighMiles': 'VisibilityHighKMs'},inplace = True)
dfWeather2.head()
```

##### 2.6 Chuyển đổi cột tầm nhìn gần từ miles sang kilomet và lưu vào cột VisibilityLowKMs trong dfWeather2

```python
dfWeather2['VisibilityLowMiles']= dfWeather2['VisibilityLowMiles'] * 1.609344
dfWeather2.rename(columns={'VisibilityLowMiles': 'VisibilityLowKMs'},inplace = True)
dfWeather2.head()
```

##### 2.7 Chuyển đổi cột sức gió từ miles/hour (WindMPH) sang kilomets/hour và lưu vào cột WindKMPH trong dfWeather2

```python
dfWeather2['WindMPH']= dfWeather2['WindMPH'] * 1.609344
dfWeather2.rename(columns={'WindMPH': 'WindKMPH'},inplace = True)
dfWeather2.head()
```

##### 2.8 Chuyển đổi cột lượng mưa từ inches sang milimet và lưu vào cột PrecipitationSumMMs trong dfWeather2
    - công thức chuyển từ Inches sang Milimet: mm = inches * 25.4
    - ví dụ chuyển từ 10 inches sang milimet: mm = 10 * 25.4 = 254 mm

```python
dfWeather2['PrecipitationSumInches']= dfWeather2['PrecipitationSumInches'] * 25.4
dfWeather2.rename(columns={'PrecipitationSumInches': 'PrecipitationSumMMs'},inplace = True)
dfWeather2.head()
```

### 5. Từ dfWeather2, liệt kê thông tin thời tiết của các ngày có nhiệt độ trung bình thấp nhất

```python
dfWeather2.sort_values('TempMediumC',ascending= True).head()
```

### 6. Từ dfWeather2, liệt kê thông tin thời tiết của các ngày có mưa, sắp giảm theo lượng mưa

```python
dfWeather2.loc[dfWeather2['PrecipitationSumMMs']>0].sort_values('PrecipitationSumMMs',ascending=False)
```

### 7. Từ dfWeather2, liệt kê thông tin thời tiết của các ngày không có mưa và có độ ẩm trung bình cao nhất

```python
dfWeather2['HumidityMeanPercent']=(dfWeather2['HumidityHighPercent'] + dfWeather2['HumidityLowPercent'])/2
dfWeather2[dfWeather2['PrecipitationSumMMs']==0].sort_values(by='HumidityMeanPercent',ascending = False)
```

## 8. Tạo series temp_medium chứa hai giá trị trung bình cột TempMediumC của các ngày có mưa và các ngày không mưa trong tháng

```python
rain_days = dfWeather2[dfWeather2['PrecipitationSumMMs'] > 0]['TempMediumC'].mean()
no_rain_days = dfWeather2[dfWeather2['PrecipitationSumMMs'] == 0]['TempMediumC'].mean()
temp_medium = pd.Series({
    "RainDays": rain_days,
    "NoRainDays": no_rain_days
})
temp_medium
```

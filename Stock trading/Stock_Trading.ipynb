```python
# Câu 1b: Đọc file stocks2.csv => đưa dữ liệu vào stocks2
stocks2 = pd.read_csv('stock_trading_data/stocks2.csv', delimiter = ',')
# Hiển thị 5 dòng dữ liệu đầu của stocks2
stocks2.head()
```

```python
# Câu 1c: Đọc file companies.csv => đưa dữ liệu vào companies
companies = pd.read_csv('stock_trading_data/companies.csv',delimiter = ',')
# Xem dữ liệu của companies
companies
```

```python
# Cho biết kiểu dữ liệu (dtype) của các cột của companies
companies.dtypes
```

```python
# Xem thông tin (info) của companies
companies.info()
```

```python
# Câu 2: Cho biết trong stocks1 có dữ liệu Null hay không?
stocks1.isnull().sum()
```

```python
stocks1[stocks1['high'].isnull()]
```

```python
stocks1[stocks1['low'].isnull()]
```

```python
# Nếu có, hãy thay thế với quy tắc sau:
# Nếu Null cột 'high' thì thay bằng giá trị max trên cột 'high' của mã chứng khoán đó
# Nếu Null cột 'low' thì thay bằng giá trị min trên cột 'low' của mã chứng khoán đó
```

```python
stocks1['high'] = stocks1['high'].fillna(stocks1['high'].max())
stocks1
```

```python
stocks1['low'] = stocks1['low'].fillna(stocks1['low'].min())
stocks1
```

```python
# Câu 3: Tạo dataframe stocks bằng cách gộp stocks1 và stocks2 theo dòng
stocks = pd.concat([stocks1,stocks2],ignore_index = 0, axis = 0)
stocks
# Xem 15 dòng dữ liệu cuối của stocks
stocks.tail(15)
```

```python
# Câu 4: Tạo dataframe stocks_companies bằng cách gộp stocks và companies
stocks_companies = pd.merge(stocks, companies, left_on = 'symbol',right_on='name',how = 'left')
stocks_companies
# Xem 5 dòng dữ liệu đầu của stocks_companies
```

```python
# Câu 5: Cho biết giá (open, high, low, close) trung bình  và volume trung bình của mỗi công ty
stocks_companies[stocks_companies['name']=='AMZN'][['open','high','low','close','volume']].mean()
```

```python
# Câu 6: Cho biết giá đóng cửa (close) trung bình, lớn nhất và nhỏ nhất ở mỗi công ty
stocks_companies.groupby('name')['close'].agg(['min', 'max', 'mean']).round(2)
```

```python
# Câu 7: Tạo cột parsed_time trong stocks_companies bằng cách đổi thời gian sang định dạng DateTime
stocks_companies['parsed_time'] = pd.to_datetime(stocks_companies['date'])
stocks_companies
# Cho biết kiểu dữ liệu của cột parsed_time
stocks_companies.dtypes
# Hiển thị 5 dòng dữ liệu đầu của stocks_companies
stocks_companies.head()
```

```python
stocks_companies['result'] = np.where(
    stocks_companies['close'] > stocks_companies['open'],
    'up',
    'down'
)
stocks_companies
```

```python

AMZN = stocks_companies[stocks_companies['symbol'] == 'AMZN']
AMZN
```

```python
# Câu 9: Vẽ biểu đồ line chart thể hiện xu hướng khối lượng giao dịch của AMZN và nhận xét
import matplotlib.pyplot as plt
plt.figure(figsize = (12,6))
plt.plot(AMZN['parsed_time'], AMZN['volume'], marker='o', linestyle='-', color='teal')
plt.title('Xu hướng khối lượng giao dịch của AMZN')
plt.xlabel('Thời gian')
plt.ylabel('Khối lượng giao dịch')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Nhận xét: Nhìn chung xu hướng khối lượng giao dịch của AMZN có biến động rồi trở lại vị trí ban đầu trong khoảng thời gian từ 1/3/2019 tới 7/3/2019. Từ ngày 1 tới ngày 4 xu hướng khối lượng giao dịch tăng từ 5.0 đến hơn 6.5. Sau đó, giảm đáng kể xuống gần 3.0 vào ngày 5 rồi lại tăng trở lại gần 5.0 vào cuối chu kì. Tóm lại xu hướng giao dịch của AMZN biến động ngắn hạn nhưng không thay đổi dài hạn. 
```

```python
# Câu 10: Vẽ biểu đồ line chart thể hiện xu hướng giá đóng cửa close của FB và nhận xét
FB = stocks_companies[stocks_companies['symbol'] == 'FB']
plt.figure(figsize=(12, 6))
plt.plot(FB['parsed_time'], FB['close'], marker='o', linestyle='-', color='darkorange')
plt.title('Xu hướng giá đóng cửa của FB theo thời gian')
plt.xlabel('Thời gian')
plt.ylabel('Giá đóng cửa (USD)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
